```python
from backtesting import Strategy, Backtest
from strategy import get_MSM_startegy
import numpy as np
import pandas as pd
```

```python
symbol = 'GESHIP'
df = pd.read_csv(f'../data/symbols/{symbol}.csv')
df['date'] = pd.to_datetime(df['date'])
df.set_index(df['date'], inplace=True)
ar_weights = np.load(f'../models/{symbol}.npy')
```

```python
data = df[['open', 'high', 'low', 'close', 'returns']]
data = data.rename(columns={'open':'Open', 'high':'High', 'low':'Low', 'close':'Close'})
```

```python
split = 0.7
train_data, test_data = data[:int(split*len(df))], data[int(split*len(df)):]
test_data = test_data.reset_index(drop=True)
```

```python
train_data
```

```python
MSM_Strategy = get_MSM_startegy(ar_weights, data)
```

```python
backtest = Backtest(data, MSM_Strategy)
backtest.run()
```

```python
backtest.plot()
```

```python
train_backtest = Backtest(train_data, MSM_Strategy)
train_backtest.run()
```

```python
train_backtest.plot()
```

```python
test_backtest = Backtest(test_data, MSM_Strategy)
test_backtest.run()
```

```python
test_backtest.plot()
```
