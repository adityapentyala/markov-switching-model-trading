```python
import pandas as pd
from scipy import stats
from statistics import mean
```

```python
df = pd.read_csv('../data/nifty_smallcap_250_ratios.csv')
df = df[df['roe']>0]
df = df[df['ev_ebitda']>0]
df = df[df['pe_ratio']>0]
df.head()
```

```python
pe_ratio_percentile = []
ev_ebitda_percentile = []
for idx in df.index:
    pe_ratio_percentile.append(stats.percentileofscore(df['pe_ratio'], df.loc[idx, 'pe_ratio'])/100)
    ev_ebitda_percentile.append(stats.percentileofscore(df['ev_ebitda'], df.loc[idx, 'ev_ebitda'])/100)

df['ev_ebitda_percentile'] = ev_ebitda_percentile
df['pe_ratio_percentile'] = pe_ratio_percentile
```

```python
df
```

```python
value_score = []

for idx in df.index:
    value_score.append(mean([df.loc[idx, 'pe_ratio_percentile'], df.loc[idx, 'ev_ebitda_percentile']]))

df['value_score'] = value_score
```

```python
df.sort_values(by='value_score', inplace=True)
df
```

```python
portfolio = df[:25]
portfolio.set_index('SYMBOL')
```

```python
portfolio.to_csv('../data/portfolio.csv', index=False)
```
